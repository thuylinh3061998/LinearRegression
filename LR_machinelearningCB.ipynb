```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
result = pandas.read_csv('E:\\nam4\\doan\\svmc\\data.csv')
print("result ban dau:\n {}".format(result))
#random ngau nhien tat ca cac ban ghi va cap nhat lai index
result=result.sample(frac=1).reset_index(drop=True);
print("result thay doi:\n {}".format(result))
      
test=result.iloc[0:20,:].reset_index(drop=True)
training=result.iloc[20:100,:].reset_index(drop=True)
print("tap test:\n {}".format(test))
print("tap training :\n {}".format(training))




```

```python
#Visualize test
X_test=np.array(test.iloc[:,:-1])
y_test=np.array(test.iloc[:,-1:])
print("X_test la: \n {}".format(X_test))
print("y_test la: \n {}".format(y_test))
plt.plot(X_test, y_test, 'ro')
plt.axis([140, 190, 35, 75])
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()

#Visualize training
X=np.array(training.iloc[:,:-1])
y=np.array(training.iloc[:,-1:])
print("X la: \n {}".format(X))
print("y la: \n {}".format(y))
plt.plot(X, y, 'ro')
plt.axis([140, 190, 35, 75])
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python


# Building Xbar 
one = np.ones((X.shape[0], 1))
print("one la: \n {}".format(one))
Xbar = np.concatenate((one, X), axis = 1)
print("Xbar la: \n {}".format(Xbar))
# Calculating weights of the fitting line 
A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T, y)
w = np.dot(np.linalg.pinv(A), b)
print('w = ', w)
# Preparing the fitting line 
w_0 = w[0]
w_1 = w[1]
x0 = np.linspace(145, 185, 2)
print("x0 la:",x0)
y0 = w_0 + w_1*x0
print("y0 la:",y0)

# Drawing the fitting line 
plt.plot(X.T, y.T, 'ro')     # data 
plt.plot(x0, y0)               # the fitting line
plt.axis([140, 190, 45, 75])
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()

```

```python
#predict voi scikit-learn
from sklearn import datasets, linear_model


# fit the model by Linear Regression
regr = linear_model.LinearRegression(fit_intercept=False) # fit_intercept = False for calculating the bias
regr.fit(Xbar, y)

# Compare two results
print( 'w  : ', regr.coef_ )
```
